```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix
from sklearn.naive_bayes import GaussianNB
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt


purchasedData = pd.read_csv('/content/Purchase_Logistic.csv')
X = purchasedData.iloc[:, [2, 3]].values
y = purchasedData.iloc[:, 4].values

scaler = StandardScaler();
X = scaler.fit_transform(X)

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=0)

cf = GaussianNB()
cf.fit(X_train, y_train)
y_pred = cf.predict(X_test)

cmat = confusion_matrix(y_test, y_pred)
print('Confusion matrix = \n', cmat)

plt.figure(1);
plt.scatter(X[:, 0], X[:, 1], c=y)
plt.suptitle('Purchase Data')
plt.xlabel('Scaled Age')
plt.ylabel('Scaled Income')
plt.grid(1, which= 'Both')
plt.axis('tight')
plt.show()

col = cf.predict_proba(X)

plt.figure(2);
plt.scatter(X[:, 0], X[:, 1], c= col[:, 1])
plt.suptitle(' Naive Bayes Purchase Data ')
plt.xlabel('Scaled Age')
plt.ylabel('Scaled Income')
plt.grid(1, which= 'Both')
plt.axis('tight')
plt.show()
```
